```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("amazon prime movies.csv")
data
```

```python
data.describe()
```

```python
data['Language'].value_counts()
```

```python
(data['IMDb Rating']=='None').value_counts()
```

```python
data['hours'] = data['Running Time'].str.extract('(\d+) h', expand=False)
data['hours'].fillna('0', inplace=True)
data['hours'].unique()
```

```python
data[['minutes',1]] = data['Running Time'].str.extract('(\d+) min|(\d+)min', expand=True).fillna(0).astype(int)
```

```python
data.loc[data[1] != 0, 'minutes'] = data[1]
data['minutes'].fillna(0, inplace=True)
data['minutes'].unique()
```

```python
data = data.drop(['Running Time', 'Plot', 1], axis=1)
data
```

```python
(data['Year of Release']).unique()
```

```python
(data['Maturity Rating']).value_counts()
```

```python
data
```

```python
data['IMDb Rating'] = data['IMDb Rating'].replace('None', np.nan)
data['IMDb Rating'] = data['IMDb Rating'].replace('         None', np.nan)
data
```

```python
data['IMDb Rating'].unique()
```

```python
data['IMDb Rating'] = data['IMDb Rating'].astype(float)
data
```

```python
data['Year of Release'] = data['Year of Release'].replace('None', 0)
data
```

```python
data['Year of Release'] = data['Year of Release'].astype(int)
data
```

```python
data['Year of Release'] = data['Year of Release'].replace(0, np.nan)
data
```

```python
data.to_csv('Amazon_Prime_edited.csv', index=False)
```

```python
data = pd.read_csv("Amazon_Prime_edited.csv")
for i in data.columns:
    print(data[i].unique())
```

```python
data['duration'] = data.apply(lambda row: '<30 mins' if row['minutes'] < 30 and row['hours'] == 0
                            else '30-60 mins' if row['minutes'] < 60 and row['hours'] == 0
                            else '1-2 hrs' if row['hours'] == 1
                            else '2-3 hrs' if row['minutes'] < 60 and row['hours'] == 2
                            else '>3 hrs', axis=1)
data['duration']
```

```python
data.to_csv('Amazon_Prime_edited.csv', index=False)
```
